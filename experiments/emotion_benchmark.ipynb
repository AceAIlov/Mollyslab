# Azumi Emotion BenchmarkThis notebook compares Azumi’s emotional responses with baseline models.

```
import requests, random
phrases = ['hello', 'I am sad', 'what is beauty?', 'good morning']
for text in phrases:
    res = requests.post('http://localhost:8000/emotion', json={'text': text}).json()
    print(f"{text} → {res['feeling']} ({res['temperature']})")
```
